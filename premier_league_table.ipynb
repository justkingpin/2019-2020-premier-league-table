```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
footie = pd.read_csv('https://raw.githubusercontent.com/footballcsv/england/master/2010s/2019-20/eng.1.csv')
footie
```

```python
# this func will split the 'ft' column into 'home' and 'away' columns

def mr_split(params):
    params['home'] = params['FT'].split('-')[0]
    params['away'] = params['FT'].split('-')[-1]
      
    return params

footie2 = footie.apply(mr_split, 1)
```

```python
# creating home winners dataframe...

home_winner = footie2[footie2['home'] > footie2['away']]
home_winner['winner'] = home_winner['Team 1']
home_winner['goals'] = home_winner['home']
home_winner
```

```python
# creating home losers dataframe...

home_loser = footie2[footie2['home'] < footie2['away']]
home_loser['loser'] = home_loser['Team 1']
home_loser['goals against'] = home_loser['away']
home_loser
```

```python
# creating away winners dataframe...

away_winner = footie2[footie2['away'] > footie2['home']]
away_winner['winner'] = away_winner['Team 2']
away_winner['goals'] = away_winner['away']
away_winner
```

```python
# creating away losers dataframe...

away_loser = footie2[footie2['away'] < footie2['home']]
away_loser['loser'] = away_loser['Team 2']
away_loser['goals against'] = away_loser['home']
away_loser
```

```python
# merging the winners dataframe & getting total goals...

merger = pd.merge(left=home_winner, right=away_winner, on=['Round', 'Date', 'winner', 'goals'], how='outer')
merger.drop(
    columns=[ 'Team 1_x', 'FT_x', 'Team 2_x', 'home_x', 'away_x',
                     'Team 1_y', 'FT_y', 'Team 2_y', 'home_y', 'away_y'],
                    inplace=True)

df = merger
df['goals'] = df['goals'].astype(int)

df = df.groupby('winner')[['goals']].apply(np.sum)
df.reset_index(inplace=True)
df.rename(
columns=
{'winner': 'Team',
'goals': 'Goals for'},
inplace=True)
df.rename(
columns={
    'Team': 'Teams'
}, inplace=True)

df
```

```python
# merging the losers dataframe & getting total against...

loss_merger = pd.merge(left=home_loser, right=away_loser, on=['Round', 'Date', 'loser', 'goals against'], how='outer')
loss_merger.drop(
    columns=[ 'Team 1_x', 'FT_x', 'Team 2_x', 'home_x', 'away_x',
                     'Team 1_y', 'FT_y', 'Team 2_y', 'home_y', 'away_y'],
                    inplace=True)
df2 = loss_merger
df2['goals against'] = df2['goals against'].astype(int)

df2 = df2.groupby('loser')[['goals against']].apply(np.sum)
df2.reset_index(inplace=True)
df2.rename(
columns=
{'loser': 'Teams'},
inplace=True)

df2
```

```python
#getting total wins...

total_wins = merger['winner'].value_counts()
wins_df = pd.DataFrame(data=total_wins)
wins_df.reset_index(inplace=True)
wins_df.rename(
columns=
    {'index': 'Teams', 
    'winner': 'Total wins'}, inplace=True
)

wins_df
```

```python
#  getting total wins & goals

df = df.merge(wins_df, on='Teams', how='outer')
df = df[['Teams', 'Total wins', 'Goals for']]
df = df.sort_values(by='Total wins', ascending=False)
df.reset_index(drop=True, inplace=True)

df
```

```python
# getting draws dataframe...

draws = footie2[footie2['home'] == footie2['away']]
draws.drop(columns = ['home', 'away'], inplace=True)
draws.reset_index(drop=True, inplace=True)
draws

total_draws = draws['Team 1'].value_counts() + draws['Team 2'].value_counts()
draws_df = pd.DataFrame(total_draws)
draws_df.reset_index(inplace=True)
draws_df.rename(columns=
               {'index': 'Teams',
               0: 'Draws'}, inplace=True
               )

draws_df
```

```python
# getting the standings...

merge1= df.merge(draws_df)
merge1.rename(columns=
             {'Total wins': 'Wins'}, inplace=True)

merge1['Loss'] = 38 - (merge1['Wins'] + merge1['Draws']) # there are 38 total matches in a season

# the func below will create a 'total' column from wins & draws

def add_up(params):
    params['Total'] =( params['Wins']*3) + params['Draws']
    return params

merge1 = merge1.groupby('Teams').apply(add_up)
final_standings = merge1.sort_values(by='Total', ascending=False)
final_standings.reset_index(drop=True, inplace=True)
final_standings
```

```python
# CAVEAT!!! 
# Chelsea had more wins, than Manutd, but the latter had more draws, and llesser losses
# In addition, Manutd had more goal difference(Goals for minus Goals against). This is why they finish above Chelsea
# Similar patterns noticed between Tottenham & Wolves, Sheffield & Burnley, Bournemouth & Watford(although they both relegated)

final_standings = final_standings.merge(df2, on='Teams', how='outer')
final_standings['GD'] =final_standings['Goals for'] - final_standings['goals against']
final_standings.rename(
columns=
{'Goals for': 'GF',
'goals against': 'GA'},
inplace=True)

keep2 = ['Teams', 'Wins', 'Draws', 'Loss', 'GF', 'GA', 'GD', 'Total']
final_standings = final_standings[keep2]
final_standings = final_standings.sort_values(by=['Total', 'GD'], ascending=[False, False])
final_standings
```
